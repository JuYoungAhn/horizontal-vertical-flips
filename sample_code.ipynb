```python
import numpy as np
import augmentation
import matplotlib.pyplot as plt
import tensorflow as tf

# tf.enable_eager_execution()
```

```python
sample_x = np.load("sample_images/sample_x.npy")
```

```python
def plot_gray_image(x):
    fig = plt.figure()
    ax = plt.imshow(x.reshape([x.shape[0], x.shape[1]]), cmap=plt.cm.gray)
    ax = plt.Axes(fig, [0, 0, 1, 1])
    plt.axis("off")
    plt.show()
```

```python
def plot_color_image(x):
    fig = plt.figure()
    ax = plt.imshow(x.reshape([x.shape[0], x.shape[1], 3]).astype(int))
    ax = plt.Axes(fig, [0, 0, 1, 1])
    plt.axis("off")
    plt.show()
```

## Gray-scale image

```python
augmented_x = augmentation.get_horizontal_vertical_flips(sample_x, ch=1)
```

```python
for i in range(0, 8):
    plot_gray_image(augmented_x[i])
```

## Color image

```python
from PIL import Image

image_obj = Image.open("sample_images/sample.jpg")
image = np.array(image_obj)
```

```python
plot_color_image(image)
```

```python
augmented_images = augmentation.get_horizontal_vertical_flips(
    image.reshape([1, image.shape[0], image.shape[1], 3]), ch=3
)
```

```python
for i in range(0, augmented_images.shape[0]):
    plot_color_image(augmented_images[i])
```
